# **Recursion and Analysis of Recursive Function**

The process in which a function calls itself directly or indirectly is called recursion and the corresponding function is
 called as recursive function. Using recursive algorithm, certain problems can be solved quite easily.


**Advantages of Recursion**


1.   Recursive functions make the code look clean and elegant.
2.   A complex task can be broken down into simpler sub-problems using recursion.
1.   Sequence generation is easier with recursion than using some nested iteration.


**Disadvantages of Recursion**
1.  Sometimes the logic behind recursion is hard to follow through. 
2. Recursive calls are expensive (inefficient) as they take up a lot of memory and time.
1.   Recursive functions are hard to debug.



Let us understand the template of a recursive function 


```
function function1(parameter ...){
  statement 1
  statement 2
  if base-condition the {
    statement 3
    statement 4
    function1(parameter...)
  }
}
```


Here the base consition tells us when the recursive should stop.

Let us understand this with an example :



````
def cal_recur(n):
  if n>0:
    k = n**2
    print(k)
    cal_recur(n-1)


cal_recur(5)
````

Now let's implement the same algorithm using loop and calculate its time complexity.

````
def cal_itr(n):
  while n>0 :
    k = n**2
    print(k)
    n = n-1


cal_itr(5)
````

For the calculation of time complexity :



Therefore we can say that the time complexity for this algorithm is  **O(n)**

**Types of Recursion**

1.   Tail Recursion
2.   Head Recursion
1.   Tree Recursion
2.   Indirect Recursion


The previous example discussed is an example of **Tail Recursion**. In Tail Recursion the function call itself and is the last statement. No statement is written after the function call. 

If a recursive function calling itself and that recursive call is the first statement in the function then it’s known as **Head Recursion**. There’s no statement, no operation before the call. The function doesn’t have to process or perform any operation at the time of calling and all operations are done at returning time. 

Let us understand this with an example code. 

````
def head_cal(n):
  if n > 0 :
    head_cal(n-1)
    k = n**2
    print(k)

head_cal(5)
````

When a function calls itself more than once then it is called **Tree Recursion**.
Let us understand it better with an example.


````
def tree_recur(n):
  if n > 0:
    tree_recur(n-1)
    k = n**2
    print(k)
    tree_recur(n-1)

tree_recur(3)
````

**Some Examples of Recursive Function**

Sum of N natural numbers

````
def summ(n):
  if n==1:
    return 1
  else:
    return summ(n-1)+n


k = summ(3)
print(k)
````

**Factorial**


````
def fact(n):
  if n==0:
    return 1
  else :
    return fact(n-1)*n


k = fact(3)
print(k)
````
